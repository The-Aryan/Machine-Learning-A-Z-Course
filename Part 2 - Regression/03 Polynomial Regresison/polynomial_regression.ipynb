```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = pd.DataFrame(dataset.iloc[:, 1].values)  # Independent Variables
Y = pd.DataFrame(dataset.iloc[:, -1].values)  # Dependent Variable
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()
linear_regressor.fit(X_train, Y_train)
```

```python
Y_pred = linear_regressor.predict(X_test)
Y_pred
```

```python
plt.scatter(X_train, Y_train, color="red")
plt.plot(X_train, linear_regressor.predict(X_train), color="blue")
plt.show()
```

```python
plt.scatter(X_test, Y_test, color="red")
plt.plot(X_test, linear_regressor.predict(X_test), color="blue")
# X_train gives the same results.
plt.show()
```
